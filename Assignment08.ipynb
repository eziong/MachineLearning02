```python
import matplotlib.pyplot as plt
import numpy as np

file_data   = "mnist_test.csv"
handle_file = open(file_data, "r")
data        = handle_file.readlines()
handle_file.close()

size_row    = 28    # height of the image
size_col    = 28    # width of the image

num_image   = len(data)
count       = 0     # count for the number of images

#
# make a matrix each column of which represents an images in a vector form 
#
list_image  = np.empty((size_row * size_col, num_image), dtype=float)
list_label  = np.empty(num_image, dtype=int)

for line in data:

    line_data   = line.split(',')
    label       = line_data[0]
    im_vector   = np.asfarray(line_data[1:])

    list_label[count]       = label
    list_image[:, count]    = im_vector
    count += 1


### Definitions of z, sigma, objective function, weight, high dimensional function
def z(theta,X):
  return g(theta,X)

def g(theta, X):
  z = 0
  for i in range(len(X)):
    z = z + theta[i]*X[i]
  return z

def sigma(z):
  return 1/(1+np.exp(-z))

def J(l,z,R):
  sum = 0
  for i in range(len(l)):
    sum = sum + l[i]*(-np.log(sigma(z[i])))+(1-l[i])*(-np.log(1-sigma(z[i])))
  return sum + R

def J_prime(_z,_l,_xi):
    return (sigma(_z)-_l)*_xi


def GD_algorithm(theta,z,l,X,alpha,Lambda):
  for i in range(len(l)):
    for j in range(len(X)):
        theta[j] = (1-alpha*Lambda)*theta[j] - alpha*w(z,l[i],X[j])
  return theta


def make_theta(X):
  ret = np.random.normal(0.0,1.0,len(X))
  return [ret for _ in range(10)]

def w(z,l,xi):
  return J_prime(z,l,xi)


def Regularization(Lambda,theta):
  sum = 0
  for theta_i in theta:
    sum = sum + theta_i*theta_i
  return sum*Lambda/2

###Initialization of theta, learning rate ,regularization and lambda
theta = make_theta(list_image)
#lr = 1e-3
lamb = 0
R = Regularization(lamb,theta[0])
ret1 = np.zeros(10)
ret2 = np.zeros(10)
lc = np.zeros(10)
X = np.transpose(list_image)

###Computing
#plot the average image
for i in range(len(X)):
    ret1[list_label[i]] = ret1[list_label[i]] + sum(X[i])
    lc[list_label[i]] = lc[list_label[i]] + 1

ret1 = ret1/lc
    
#find theta by Gradient Decent Algorithm
#for i in range(len(X)): # 10,000
#    zet = z(theta[list_label[i]],X[i])
#    theta[list_label[i]] = GD_algorithm(theta[list_label[i]],zet,[list_label[i]],X[i],lr,R)

#Present the output of the neural network with random weights
for i in range(len(X)): # 10,000
    zet = z(theta[list_label[i]],X[i])
    ret2[list_label[i]] = ret2[list_label[i]] + sigma(zet)

ret2 = ret2/lc

#Plot the average image
print("#Plot the average image")
print(ret1[0]," ",ret1[1]," ",ret1[2]," ",ret1[3]," ",ret1[4]," ")
print(ret1[5]," ",ret1[6]," ",ret1[7]," ",ret1[8]," ",ret1[9]," ")
print()

#Present the output of the neural network with random weights
print("#Present the output of the neural network with random weights")
print(ret2[0]," ",ret2[1]," ",ret2[2]," ",ret2[3]," ",ret2[4]," ")
print(ret2[5]," ",ret2[6]," ",ret2[7]," ",ret2[8]," ",ret2[9]," ")
```
